## Counter

```python
from collections import Counter
```

```python
Counter?
```

```python
c1 = Counter("Ankit")                     # It'll count all the letters in the string and store them
print(c1)
c2 = Counter(['a', 'b', 'c', 'c', 'a'])   # Will count duplicate items
print(c2)
c3 = Counter({'a':2, 'b':3, 'c': 5})      # Will store it as it is
print(c3)
c4 = Counter(a=2, b=3, c=5)               # Same as 5th line
print(c4)

print(type(c1))                           # It's a seperate object of class -> collections.Counter
```

#### Accessing Values

```python
print(c4['b'])                            # Gets value related to the specified key as in dictionary
print(c4['d'])                            # Unlike in dictionary it won't throw an error
```

#### Useful Methods

```python
print(list(c1.elements()))                # get all items
print(list(c4.elements()), "\n")

print(c4.most_common(1))                  # gets 'the' most common elements
print(c4.most_common(2), "\n")            # gets first 2 most common element
```

```python
x = Counter(a=4, b=8, c=-5, d=3)
y = ['a', 'b', 'c', 'a']

x.subtract(y)                             # subtracts y from x
print("Substracted value:  ", x)

x.update(y)                               # adds y in x
print("Updated value:      ", x)

x.clear()                                 # clears entire counter object
print("Cleared:            ", x)
```

```python
x = Counter(a=4, b=8, c=-5, d=3)
y = Counter(['a', 'b', 'c', 'a'])

print("x + y =", x + y)                   # Anything <=0 will be discarded & rest will be added  
print("x - y =", x - y)                   # Anything <=0 will be discarded & rest will be substracted
```

#### Union ('|') and Intersection ('&')

```python
print("Union (|):        ", x | y)        # Picks elements with max values out of both counters
print("Intersection (&): ", x & y)        # Picks elements with min values out of both counters
```

## NamedTuple

```python
from collections import namedtuple
```

```python
p = namedtuple('Point', 'x ay z')         # Tuple name = 'Point', 'x y z' will be broken as seperate parameter by spaces
newP = p(3, 4, 5)                         # You must pass equal no. of args to parameters (which are x, y, z)
print(newP, '\n')

p1 = namedtuple('Point', ['a', 'bd', 'c'])# The second parameter can be any iterable
newP1 = p1(1, 3, 7)                       
print(newP1, '\n')

newP2 = p._make(['m', 'n', 'o'])          # To assign values from list use _make() method else it'll throw error
print(newP2)
```

#### Accessing values

```python
print(newP1[0], newP1[1], newP1[2])
print(newP1.a, newP1.bd, newP1.c, '\n')

print(newP1._asdict())                    # Prints in dictionary format
print(newP1._fields)                      # Returns tuple composed of fields
```

```python
# Similar to tuples you cannot change the values in namedtuples, but you may reassign them
newP1._replace(a=10)                     # Returns a new tuple similar to newP1 except it changes a=10
print(newP1)                             # No change in newP1 as namedtuple is immutable

newP1 = newP1._replace(a=10)             # Reassign
print(newP1)                             # Now you'll see the change
```
