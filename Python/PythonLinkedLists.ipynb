# Singly Linked List

## Class for Creating Singly Linked List Node 

```python
class Node:
    
    def __init__(self, data=None):
        self.data = data
        self.next = None
    
    def __str__(self):
        return str(self.data)
        
```

## Class for creating Singly Linked List 
LinkedList simply interconnects nodes

```python
class LinkedList:
    
    def __init__(self, data=None):
        self.head = None
        self.tail = None
        
        if data is not None:
            self.addMultipleNodes(data)
            
            
    def __len__(self):
        count = 0
        curr = self.head
        while curr:
            count += 1
            curr = curr.next
        
        return count
    
        
    #insertion   
    def addNode(self, data):
        if self.head is None:                    #If linked list is empty self.tail == None and will not have 
            self.head = self.tail = Node(data)   #.next attribute and will throw error if self.tail.next is called
        else:
            self.tail.next = Node(data)
            self.tail = self.tail.next
    
    
    def addMultipleNodes(self, data):
        for i in data:
            self.addNode(i)
            
            
    def addHead(self, data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode
        
        
    #addNode() method will work as addTail method
    
    
    def inBetween(self, data, idx):
        i = 0
        curr = self.head
        while i<idx-1:                             #Find node just before the target index 'idx'
            curr = curr.next
            i+=1
        
        newNode = Node(data)
        newNode.next = curr.next
        curr.next = newNode
            
    
    #generic method to handle all
    def push(self, data, idx=None):
        if idx is None:                            #When no index is passed append at the end
            idx = len(self)
        elif idx < 0:
            print("Invalid Index. Reverse indexing is not supported")
            return
        
        if idx > (len(self)-1):                    #If Linked List is empty or if index is out of bound
            
            if type(data) in [list, tuple, set, dict]:
                self.addMultipleNodes(data)
                return
            
            self.addNode(data)
            return
        
        newNode = Node(data)
        
        if idx == 0:                               #Insert new head in non empty Linked List
            newNode.next = self.head
            self.head = newNode
            return
            
        i = 0
        curr = self.head
        while i < idx-1:                             #Find node just before idx
            curr = curr.next
            i += 1
        
        newNode.next = curr.next
        curr.next = newNode
        
        if newNode.next is None:
            self.tail = newNode
        
        
    #deletion
    def pop(self, data, idx=None):
        curr = self.head
        
        if curr is None:
            print("Linked List is empty!")
            return
        
        if curr.data == data:                      #When target node is head node
            self.head = curr.next
            return
        
        prev = curr
        curr = curr.next
        while curr != None:
            if curr.data == data:
                prev.next = curr.next
                if prev.next is None:
                    self.tail = prev
                return
            prev = curr
            curr = curr.next
            
        if curr is None:                            #Target not found
            print("Element does not exist!")
          
    
    #traversal
    def printAll(self):
        if self.head is None:
            #raise Exception("Empty Linked List!")
            print("Empty Linked List!")
        else:
            curr = self.head
            while curr:
                print(curr, end=" ", flush=True)
                curr = curr.next
                
                
    def getAll(self):
        if self.head is None:
            #raise Exception("Empty Linked List!")
            print("Empty Linked List!")
        else:
            ans = []
            curr = self.head
            while curr:
                ans.append(curr.data)
                curr = curr.next
            return ans
        
     
    def reverse(self):
        prev = None
        curr = self.head
        self.tail = self.head
        while curr:
            nextNode = curr.next
            curr.next = prev
            prev = curr
            curr = nextNode
        self.head = prev
```

```python
L1 = LinkedList()            #Empty Linked List
L2 = LinkedList([5, 5])
```

```python
print(L1)
print(L2)
```

```python
L1.printAll()    #Will print "Empty Linked List!"
```

```python
L2.printAll()    #Will print all node data
```

```python
L1.push([1,4,7])
L1.printAll()
```

```python
print("Initial:", L2.getAll())

L2.pop(50)                            #Pop at start
print("Poped: ", L2.getAll())

L2.pop(60)                            #Pop in between
print("Poped: ", L2.getAll())

L2.pop(78)                            #Pop at end
print("Poped: ", L2.getAll())

L2.pop(5)                             #When multiple elements are present first occurence is poped
print("Poped: ", L2.getAll())

L2.pop(560)                           #When element is absent 
print("Not Poped: ", L2.getAll())
```

# Doubly Linked List

## Class for creating Doubly Linked List Node

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
    
    def __str__(self):
        return str(self.data)
```

## Class for  creating Doubly Linked List

```python
class DoublyLinkedList:
    def __init__(self, data=None):
        self.head = None
        self.tail = None
        
        if data is not None:
            self.addMultipleNodes(data)
            
            
    def __len__(self):
        count = 0
        curr = self.head
        while curr:
            count += 1
            curr = curr.next
        
        return count
    
    
    #insertion
    def addNode(self, data):
        if self.head is None:                   #If Linked List is empty
            self.head = self.tail = Node(data)
        else:
            newNode = Node(data)
            self.tail.next = newNode
            newNode.prev = self.tail
            self.tail = newNode
    
    
    def addMultipleNodes(self, data):
        for i in data:
            self.addNode(i)
    
    
    def push(self, data, idx=None):
        l = len(self)
        if idx is None or idx >= l:                         #When no index is passed append at the end
            self.addNode(data)
            return
        elif idx < 0:
            print("Invalid index. Reverse indexing is not supported.")
            return
        
        #If program comes here it means Linked List is not empty and index is within range
        newNode = Node(data)
        
        if idx == 0:                            #Insert node at head (non empty Linked List)
            self.head.prev = newNode
            newNode.next = self.head
            self.head = newNode
            return

        back = l-idx
        if back <= idx:                         #------Traverse backwards------- 
            i = 1
            curr = self.tail
            while i<back:                     #Find node just before idx
                curr = curr.prev
                i += 1
            newNode.prev = curr.prev
            newNode.next = curr
            curr.prev = newNode
            curr = newNode.prev
            curr.next = newNode
        else:
            i = 0                               #Insert node in between
            curr = self.head
            while i<idx-1:                      #Find node just before idx
                curr = curr.next
                i += 1
            newNode.next = curr.next
            newNode.prev = curr
            curr.next = newNode
            curr = newNode.next
            curr.prev = newNode
    
    
    #deletion
    def pop(self, data):
        curr = self.head
        
        if curr is None:
            print("Linked List is empty!")
            return
        
        if curr.data == data:                   #If target node is head node
            self.head = curr.next
            self.head.prev = None
            return
        
        #prev = curr
        curr = curr.next
        while curr != None:
            if curr.data == data:
                back = curr.prev
                front = curr.next
                back.next = curr.next
                if front is None:
                    self.tail = back
                else:
                    front.prev = curr.prev
                return
            curr = curr.next
            
        if curr == None:                         #Target not found
            print("Element does not exist!")
    
    
    #traversal
    def printAll(self):
        if self.head is None:
            #raise Exception("Empty Linked List!")
            print("Empty Linked List!")
        else:
            curr = self.head
            while curr:
                print(curr, end=" ", flush=True)
                curr = curr.next
    
    def getAll(self):
        if self.head is None:
            #raise Exception("Empty Linked List!")
            print("Empty Linked List!")
        else:
            ans = []
            curr = self.head
            while curr:
                ans.append(curr.data)
                curr = curr.next
            return ans
    
    
    def reverse(self):
        curr = self.tail
        ans = []
        while curr:
            ans.append(curr.data)
            curr = curr.prev
        return ans
    
    
    def reverseAll(self):
        curr = self.head
        self.tail = self.head
        ahead = curr.next
        curr.next = None
        curr.prev = ahead
        while ahead:
            ahead.prev = ahead.next
            ahead.next = curr
            curr = ahead
            ahead = ahead.prev
        self.head = curr
```

```python
DL1 = DoublyLinkedList()
DL2 = DoublyLinkedList([5, 5])
```
